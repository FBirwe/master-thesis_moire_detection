```python
from PIL import Image
import numpy as np
from tqdm.auto import tqdm
import json
```

```python
from raster_image_processor import DOT_SHAPE, RasterImageProcessor
```

```python
dotenv = load_dotenv()
```

```python
source_images = [
    (jpg_path.name.replace(jpg_path.suffix,''), Image.open(jpg_path))
    for jpg_path in dotenv['PATTERN_DIR'].glob('./*.jpg')
] + [
    (
        f'plain_{i}',
        Image.fromarray((np.ones((6000,6000)) * 255 * (i / 10)).round().astype('uint8'))
    )
    for i in range(1,10)
]
```

```python
for source_img_name, source_img in tqdm(source_images):
    for dot_shape in DOT_SHAPE:
        rip = RasterImageProcessor( dot_shape, 16 )
        img_out = rip.screen_image( source_img, 600, 2400 )
        out_path = dotenv['PATTERN_DIR'] / f'{ source_img_name }.{ dot_shape.name }.tif'
        img_out.save(
            out_path,
            dpi=(2400,2400),
            compression="tiff_lzw"
        )
```

# Pattern als JSON ablegen

```python
available_patterns = [
    tiff_path.name.replace(tiff_path.suffix,'') for tiff_path in
    dotenv['PATTERN_DIR'].glob('./*.tif')
]
```

```python
with (dotenv['PATTERN_DIR'] / 'available_patterns.json').open('w') as json_file:
    json.dump(available_patterns, json_file)
```
